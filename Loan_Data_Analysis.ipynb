```python
def security(a):
  if (a['Securities Account'] == 1) & (a['CD Account']==1):
    return "Holds Securities and Deposits"
  if(a['Securities Account']==0) & (a['CD Account']==0):
    return "Does not hold Securities and Deposits"
  if (a['Securities Account']==1) & (a['CD Account']==0):
      return "Holds Securities only"
  if (a['Securities Account']==0) & (a['CD Account']==1):
      return "Holds Deposits only"
```

```python
newdata['Account_holder_category'] = newdata.apply(security, axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def create_kdeplot(col1, col2, label1, label2, title):
  """Creates a KDE plot with two distributions."""
  # Filter data for each KDE plot
  data1 = newdata[newdata[col1] == 1][col2]
  data2 = newdata[newdata[col1] == 0][col2]

  # Check if data is empty before plotting
  if not data1.empty:
    sns.kdeplot(data1, label=label1)
  if not data2.empty:
    sns.kdeplot(data2, label=label2)

  plt.legend()
  plt.title(title)

# Call the function with appropriate columns
create_kdeplot('Personal Loan', 'Income', 'Income with personal loan', 'Income with No personal Loan', 'personal loan Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def create_kdeplot(filter_col, plot_col, label1, label2, title):
  """Creates a KDE plot with two distributions."""
  # Filter data for each KDE plot
  data1 = newdata[newdata[filter_col] == 1][plot_col] # Filter by 'Personal Loan', plot 'CCAvg'
  data2 = newdata[newdata[filter_col] == 0][plot_col] # Filter by 'Personal Loan', plot 'CCAvg'

  # Check if data is empty before plotting
  if not data1.empty:
    sns.kdeplot(data1, label=label1)
  if not data2.empty:
    sns.kdeplot(data2, label=label2)

  plt.legend()
  plt.title(title)

# Call the function with appropriate columns
create_kdeplot('Personal Loan', 'CCAvg', # Filter by 'Personal Loan', plot 'CCAvg'
     'Credit card avg with personal loan', # Label for personal loan group
     'Credit card avg with no personal loan', # Label for no personal loan group
     'Credit card avg Distribution')
plt.show()
```
